```python
label_encoders = {} 
categorical_columns = ["Model","Colour","Processor_"]

for col in categorical_columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

print(df)
```

```python
num_cols = df.select_dtypes(include=[np.number]).columns
```

Remove Outliers from dataset

```python
def remove_outliers_iqr(df,columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5*IQR
        upper_bound = Q3 + 1.5*IQR
        df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
    return df
```

```python
df_clean = remove_outliers_iqr(df, num_cols)
print("Original shape:",df.shape)
print("Cleaned shape:",df_clean.shape)
```

1. Linear Regression Model

```python
from sklearn.model_selection import train_test_split

x = df_clean.drop('Prize',axis=1)
y = df_clean['Prize']

x_train, x_test , y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Fitting of linear regression model
lr_model = LinearRegression()
lr_model.fit(x_train, y_train)
```

```python
df.head()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
y_pred = lr_model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R^2: {r2}")
```

```python
lr_model.score(x_test, y_test)
```

```python
lr_model.score(x_train, y_train)
```

```python
x.columns
```

```python
coefficients = lr_model.coef_
features = x_train.columns

importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': abs(coefficients) # use abs() to rank by strength
}).sort_values(by='Importance', ascending=False)
```

```python
#Plot feature importances
plt.figure(figsize=(10, 6))
sns.barplot(x='Importance', y='Feature', data=importance_df)
plt.title(f'Top Features Affecting Mobile Price (R² Score: {r2:.2f})')
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

*Suprisingly, Mobile Height is shown as the top feature influencing mobile price.
*As expected, RAM is one of the top predictors. Higher RAM usually indicates better performance and higher-end device.
*Front Camera also has a noticeable influence , indicating that selfie camera specs affect price.

2. Random Forest Model

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
rf_model = RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
rf_model.fit(x_train,y_train)
```

```python
y_pred = rf_model.predict(x_test)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mse**0.5
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R^2: {r2}")
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Prices")
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], color='red')
plt.tight_layout()
plt.show()
```
